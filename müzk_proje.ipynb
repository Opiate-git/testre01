```python
import json
import random 


class SongNode:
    def __init__(self, title):
        self.title = title
        self.next = None

class Playlist:
    def __init__(self):
        self.head = None
        self.current = None

    def add_song(self, title):
        new_node = SongNode(title)
        if not self.head:
            self.head = new_node
            new_node.next = new_node
            self.current = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def delete_song(self, title):
        if not self.head:
            print("Çalma listesi boş.")
            return
        temp = self.head
        prev = None
        while True:
            if temp.title == title:
                if prev:
                    prev.next = temp.next
                else:
                    if temp.next == self.head:
                        self.head = None
                        self.current = None
                        print(f"Silinen şarkı: {title}")
                        return
                    else:
                        last = self.head
                        while last.next != self.head:
                            last = last.next
                        self.head = self.head.next
                        last.next = self.head
                if self.current == temp:
                    self.current = self.head
                print(f"Silinen şarkı: {title}")
                return
            prev = temp
            temp = temp.next
            if temp == self.head:
                break
        print(f"Şarkı '{title}' bulunamadı.")

    def play_next(self):
        if not self.current:
            print("Çalma listesi boş.")
            return
        print(f"Şu an çalıyor: {self.current.title}")
        self.current = self.current.next

    def play_previous(self):
        if not self.current:
            print("Çalma listesi boş.")
            return
        temp = self.head
        prev = None
        while temp.next != self.current:
            temp = temp.next
        self.current = temp
        print(f"Önceki şarkı çalınıyor: {self.current.title}")

    def display_playlist(self):
        if not self.head:
            print("Çalma listesi boş.")
            return
        print("Çalma listesi:")
        temp = self.head
        while True:
            print(f"- {temp.title}")
            temp = temp.next
            if temp == self.head:
                break

    def save_to_file(self, filename):
        if not self.head:
            print("Çalma listesi boş - kaydedilecek bir şey yok.")
            return

        songs = []
        temp = self.head
        while True:
            songs.append(temp.title)
            temp = temp.next
            if temp == self.head:
                break

        current_song = self.current.title if self.current else None

        data = {
            'songs': songs,
            'current_song': current_song
        }

        try:
            with open(filename, 'w') as f:
                json.dump(data, f)
            print(f"Çalma listesi {filename} dosyasına kaydedildi.")
        except Exception as e:
            print(f"Kaydetme hatası: {e}")

    def load_from_file(self, filename):
        try:
            with open(filename, 'r') as f:
                data = json.load(f)

            self.head = None
            self.current = None

            for song in data['songs']:
                self.add_song(song)

            if data['current_song']:
                temp = self.head
                while True:
                    if temp.title == data['current_song']:
                        self.current = temp
                        break
                    temp = temp.next
                    if temp == self.head:
                        break

            print(f"Çalma listesi {filename} dosyasından yüklendi.")
        except FileNotFoundError:
            print(f"{filename} dosyası bulunamadı.")
        except Exception as e:
            print(f"Yükleme hatası: {e}")

    def shuffle_playlist(self):
        if not self.head or self.head.next == self.head:
            print("Çalma listesinde sıfır veya bir şarkı var; karıştırmaya gerek yok.")
            return

        songs = []
        temp = self.head
        while True:
            songs.append(temp.title)
            temp = temp.next
            if temp == self.head:
                break

        random.shuffle(songs)

        self.head = None
        self.current = None
        for title in songs:
            self.add_song(title)

        print("Çalma listesi karıştırıldı.")

def run_menu():
    pl = Playlist()
    while True:
        print("\nMenü:")
        print("1. Şarkı ekle")
        print("2. Şarkı sil")
        print("3. Sonraki şarkıyı çal")
        print("4. Önceki şarkıyı çal")
        print("5. Çalma listesini göster")
        print("6. Listeyi dosyaya kaydet")
        print("7. Dosyadan listeyi yükle")
        print("8. Listeyi karıştır (shuffle)")
        print("9. Çıkış")
        choice = input("Seçiminiz: ")

        if choice == '1':
            title = input("Eklemek istediğiniz şarkının adı: ")
            pl.add_song(title)
        elif choice == '2':
            title = input("Silmek istediğiniz şarkının adı: ")
            pl.delete_song(title)
        elif choice == '3':
            pl.play_next()
        elif choice == '4':
            pl.play_previous()
        elif choice == '5':
            pl.display_playlist()
        elif choice == '6':
            filename = input("Kaydetmek için dosya adı: ")
            pl.save_to_file(filename)
        elif choice == '7':
            filename = input("Yüklemek için dosya adı: ")
            pl.load_from_file(filename)
        elif choice == '8':
            pl.shuffle_playlist()
        elif choice == '9':
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçim.")

run_menu()
```
